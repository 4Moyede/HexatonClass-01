```python
airbnb_select = airbnb[['room_type','neighborhood','reviews']]
airbnb_room = airbnb_select.drop_duplicates(['room_type'],keep='first')
room_dict=airbnb_room.to_dict()
del room_dict['neighborhood']
del room_dict['reviews']

airbnb_neighborhood = airbnb_select.drop_duplicates(['neighborhood'],keep='first')
neighborhood_dict=airbnb_neighborhood.to_dict()
del neighborhood_dict['room_type']
del neighborhood_dict['reviews']

airbnb_reviews = airbnb_select.drop_duplicates(['reviews'],keep='first')
reviews_dict=airbnb_reviews.to_dict()
del reviews_dict['neighborhood']
del reviews_dict['room_type']

print('This is \'room_type\' and neighborhood \'s value :')
print()
print(room_dict)
print(neighborhood_dict)
print(reviews_dict)
```

```python
airbnb_filter = airbnb[(airbnb['price'] >= 250) & (airbnb['neighborhood'] == 'Downtown') & (airbnb['reviews'] >= 5)]

print('''Start Filtering Job
The number of house : 
''',len(airbnb_filter.index) )

airbnb_filter
```
